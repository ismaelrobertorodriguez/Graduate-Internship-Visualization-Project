```python
#url = 'https://www.sbir.gov/api/awards.json?'
#req = urllib.request.Request(url)

#awardReq = urllib.request.Request(url)

#r = urllib.request.urlopen(awardReq).read()
#cont = json.loads(r.decode('utf-8'))



awardDataFull = []
site = 'https://www.sbir.gov/api/awards.json?start=%s'
for page in range(0, 180000, 100):
    url = site % page
    print(url)
    awardReq = urllib.request.Request(url)
    #r = urllib.request.Request(url)
    #cont= json.loads(r.decode('utf-8'))
    r = urllib.request.urlopen(awardReq).read()
    cont = json.loads(r.decode('utf-8'))
    for i in cont:
        awardDataFull.append([i['year'],i['agency'],i['program'],i['research_institution'],i['firm']])
```

```python
awardsDF = pd.DataFrame(awardDataFull)
```
